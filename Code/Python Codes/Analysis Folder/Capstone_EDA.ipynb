```python
# Set style and context for plots
sns.set_style('whitegrid')
sns.set_context('paper')

# Create a list of the fields to plot
fields = ['Timetable_Schedules_score','Section_Allocation_score', 'Class_Hours_Score', 'Course_Contents_Score', 'Afterclass_Workload_score']

# Loop through the list and create a histogram for each field
for field in fields:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=Base_Survey_Table, x=field, kde=False, bins=10, color='lightblue')
    plt.title(field + ' Distribution')
    plt.xlabel(field + ' Score')
    plt.ylabel('Count')
    
    # Set xticks to show the rating scale of 1 to 5
    plt.xticks([1, 2, 3, 4, 5])
    
    # Add text to show the rating scale of 1 to 5
    plt.text(0.9, 0.9, 'Rating Scale: 1 - Least, 5 - Max',
             transform=plt.gca().transAxes,
             fontsize=10, verticalalignment='top')

    plt.grid(False)
    plt.show()
```
